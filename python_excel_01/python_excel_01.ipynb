### 一、创建新的工作簿
#### 并在工作簿里面创建工作表，写入内容并保存

```python
# python中用来处理excel的包有哪些？
# 1、是openpyxl（官方地址）、openpyxl只支持xlsx格式。
```

```python
# 导入需要的包
from openpyxl.workbook import Workbook
from openpyxl.writer.excel import ExcelWriter
# 这个模块是转换数字到字母的
from openpyxl.utils import get_column_letter
```

```python
# 创建工工作簿
wb = Workbook()
```

```python
# 创建第1个工作表，并写入内容
ws = wb.worksheets[0]
ws.title = "range names"
ws.cell("C1").value = '哈哈'
```

```python
# 创建第2个工作表，并写入内容
wc = wb.create_sheet("range names 2")
# wc = wb.worksheets[1]
wc.title = "range names 2"
wc.cell("A1").value='牛逼'
```

```python
# 对工作簿命名并保存
wb.save(filename ="python_excel_01_01.xlsx")
```

### 二、利用Pandas读取Excel工作表，修改并保存
#### 读入进去pandas之后就是按照DataFrame方式处理，这样就很强大，处理完之后直接另外保存就可以

```python
# 导入需要的包。先不导入这么多，后面需要什么导入什么。
import pandas as pd 
```

```python
df1 = pd.read_excel("us_baby_names.xlsx", header=None) 
# 这里的header=None用来控制不让第一行作为列名
```

```python
# 查看数据
df1.head()
```

```python
# 查看有多少行和列
df1.shape
```

```python
# 指定第一行为列名。（因为我们读取的时候指定了第一行不为列名，所以默认列名回事0,1,2这样。现在我们知道应该将第一行设为列名的）
df1.columns = df1.iloc[0]
df1.head()
```

```python
df1.shape
# 可以看到指定第一列名列的后，并不是直接将第一行移上去，而是复制第一行的内容为列名。故行数不变
```

```python
# 删除指定行。注意这里的行的索引是从0开始的,这里删除前面1行数据
df2 = df1.drop([0])
df2.head()
# 发现删除的1行已经没有了。但是索引Index不会自动更新
# 注意这里也可以按照df2.drop(1,inplace = True)这种删除方式，inplace=True表示在源数据表直接删除
```

```python
# 查看删除后的行和列数。与删除前相比少了1行
df2.shape
```

```python
# 删除指定列,axis = 1来控制删除列。删除多列或者一列都是一样的。比如我们先把第一列Id列删除不要了
df3 = df2.drop(["Id"],axis=1)
df3.head()
# 发行第一列已经没有了
```

```python
# 查看下完整数据是怎么样的
df3
```

#### 数据过滤。筛选指定条件的数据

```python
# 参考链接：https://www.jianshu.com/p/4a2ecf65e3ea
# https://chf2012.github.io/2018/02/06/%E8%BD%AF%E4%BB%B6%E5%BA%94%E7%94%A8_%E7%A8%8B%E5%BA%8F%E7%BC%96%E7%A8%8B/Python/Python_%E4%B8%93%E9%A2%98%E6%80%BB%E7%BB%93/Python_%E6%95%B0%E6%8D%AE%E5%A4%84%E7%90%86_pandas/
```

```python
captalizer = lambda x: x[0:1]
# 这里是截图Name列的首字母
```

```python
# 这里是筛选首字母为E、S或A的行。注意学习apply的用法
df4 = df3.loc[(df3["Name"].apply(captalizer) == 'E') | (df3["Name"].apply(captalizer) == 'S') | (df3["Name"].apply(captalizer) == 'A' )]
df4
```

```python
df5 = df4.sort_values(by=["Name","Count"] , ascending=True)
# ascending=True,默认也是True
df5
```

```python
# 我们看到Count列并不是按照升序排列的。查看下数据类型
df5.dtypes
```

```python
df5['Count'] = df5['Count'].astype('int')
```

```python
df5.dtypes
```

```python
df6 = df5.sort_values(by=["Name","Count"] , ascending=True)
df6
# 不知道这里怎么回事，看到排序对Count并没有起作用。其实是起作用了的，要看具体相同首字母里面看
```

#### 如何选出某一列的某些值所在的行

```python
df7 = df6[df6['Name'].isin(['Alexis','Ashley','Anna','Ava'])]
df7
```

```python
df8 = df6[df6['Name'].isin(['Sarah','Sierra','Sophia','Sydney'])]
df8
```

```python
# 将筛选出来的数据表聚合在一起
# 当列数及列名一致的时候，这里就涉及到DataFrame的操作。Pandas中DataFrame数据合并、连接（concat、merge、join）
df9 = df7.append(df8)
df9
```

```python
df9['Lover'] = ""
df9.head()
# 新增加一列是加在数据表的最后一列
# python 处理Excel电子表格：https://blog.csdn.net/tanggangzuiniu12/article/details/79197661
```

```python
# df.insert(idx, col_name, value)
# 将新增加一列移动到第4列
df9.insert(3,'Lover',df9.pop('Lover'))
df9
```

```python
df5.to_excel('python_excel_01_02.xlsx',index=False) # 设置index=False不导入索引
# 直接可以将改动进行另存为文件
```
